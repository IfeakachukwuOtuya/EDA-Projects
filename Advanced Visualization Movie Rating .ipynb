```python

sns.set_style('darkgrid')
f, axes = plt.subplots(2,2, figsize=(15,15))
k1 = sns.kdeplot(data=data, x='BudgetMillion', y='AudienceRatings', ax=axes[0,0])
k2 = sns.kdeplot(data=data, x='BudgetMillion', y='AudienceRatings', ax=axes[0,1])

k1.set(xlim=(-20.160))
k2.set(xlim=(-20.160))

z = sns.violinplot(data=data[data.Genre == 'Drama'], x='Year', y='CriticRatings', ax=axes[1,0])
k4 = sns.kdeplot(data=data, x='BudgetMillion', y='AudienceRatings', shade=True, shade_lowest=False, cmap='Greens_r')
k4b =sns.kdeplot(data=data, x='CriticRatings', y='AudienceRatings', cmap='BrBG', ax=axes[1,1])

plt.tight_layout()
plt.show()
```

```python
# How can you style your dashboard using different color map
# Python is not vectorize programming language
# Building dashboards (dashboard - combination of chats)

sns.set_style('dark',{'axes.facecolor':'white'})
f, axes = plt.subplots(2,2, figsize=(15,15))

#plot [0,0]
k1 = sns.kdeplot(data=data, x='BudgetMillion', y='AudienceRatings', shade=True, shade_lowest=False, cmap='YlGnBu_r', ax=axes[0,0])
kb1 = sns.kdeplot(data=data, x='BudgetMillion', y='AudienceRatings', cmap='cool', ax=axes[0,0])

#plt [0,1]

k2 = sns.kdeplot(data=data, x='BudgetMillion', y='CriticRatings', shade=True, shade_lowest=False, cmap='summer', ax=axes[0,1])
k2b = sns.kdeplot(data=data, x='BudgetMillion', y='AudienceRatings', cmap='cividis', ax=axes[0,1])

#plot [1,0]

z = sns.violinplot(data=data[data.Genre == 'Drama'], x='Year', y='CriticRatings', ax=axes[1,0])

#plot [1,1]
k4 = sns.kdeplot(data=data, x='CriticRatings', y='AudienceRatings', shade=True, shade_lowest=False, cmap='Grays', ax=axes[1,1])
k4b = sns.kdeplot(data=data, x='CriticRatings', y='AudienceRatings', cmap='CMRmap_r', ax=axes[1,1])

k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))

plt.show()
```
